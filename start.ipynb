```python
import subprocess, sys, os, shutil

subprocess.run([sys.executable, "-m", "pip", "install", "gradio", "-q"])

if not os.path.exists("assets"):
    try:
        os.symlink("common/assets", "assets")
    except OSError:
        shutil.copytree("common/assets", "assets", dirs_exist_ok=True)

subprocess.Popen([sys.executable, "gradio_app.py"])
print("Open http://localhost:7860")
```
