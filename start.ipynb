## Gradio Web UI 一键启动

运行下方单元格启动图形化界面：

````python
import subprocess
import sys
import time
import webbrowser
from IPython.display import display, HTML, clear_output

# Kill any existing gradio process
import os
os.system("taskkill /f /im python.exe 2>nul || pkill -f gradio_app 2>/dev/null; exit 0")
time.sleep(1)

# Start Gradio app in background
proc = subprocess.Popen(
    [sys.executable, "gradio_app.py"],
    stdout=subprocess.PIPE,
    stderr=subprocess.STDOUT,
)

# Wait for it to start
time.sleep(3)

clear_output(wait=True)
display(HTML("""
<div style='padding:20px; background:#e8f5e9; border-radius:8px; font-size:16px;'>
  <h3>✅ Gradio UI 已启动!</h3>
  <p>请点击下方链接打开界面:</p>
  <a href='/proxy/7860' target='_blank' 
     style='display:inline-block; padding:12px 24px; background:#1976d2; color:white; 
            text-decoration:none; border-radius:6px; font-size:18px; font-weight:bold;'>
    🚀 打开 Gradio 界面
  </a>
  <p style='margin-top:12px; color:#555;'>如果按钮不工作，尝试直接访问: <b>http://localhost:7860</b></p>
</div>
""")
````
